```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Marks': [85, 92, 78, 90]
}

df = pd.DataFrame(data)
print(df)
print('................')
#setting custom index as name
df.set_index('Name', inplace=True)
print(df)
print('.................')
#access a row using the new index
print(df.loc[['Alice']])

#reset index back to default
df.reset_index(inplace=True)
print(df)

#create a dataframe with range index
new_df = pd.DataFrame(data, index=pd.RangeIndex(100, 100 + len(data['Name']) * 10, 10))
print(new_df)

#
```

```python
#exercise
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [24, 30, 35, 40],
        'City': ['New York', 'Toronto', 'London', 'Sydney']}

df = pd.DataFrame(data)
print(df)

#1. Display only the Name and City columns.
new_df = df[['Name', 'City']]
print(new_df)

#2. Select and print the rows where Age is greater than 30.
filter_df = df[df['Age'] > 30]
print(filter_df)

#1. Add a new column named Salary with values [70000, 80000, 120000, 95000].
df['Salary'] = [70000, 80000, 120000, 95000]
print(df)

#2. Delete the City column.
df.drop('City', axis=1, inplace=True)

#3 3. Goal: Display the updated DataFrame.
print(df)
```
